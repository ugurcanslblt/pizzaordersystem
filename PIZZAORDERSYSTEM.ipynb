```python
import csv
import datetime

#menüyü menu.txt dosyasına yazdırır
menu_file = open("menu.txt", "w")
menu_file.write("_ MENU _")
menu_file.write("\nPizza Bases:")
menu_file.write("\n1. Pizza capricciosa")
menu_file.write("\n2. Pizza Margherita")
menu_file.write("\n3. Hawaiian pizza")
menu_file.write("\n4. Seafood pizza")
menu_file.write("\n")
menu_file.write("\nPizza Sauces:")
menu_file.write("\n1. Olives")
menu_file.write("\n2. Mushrooms")
menu_file.write("\n3. Goat cheese")
menu_file.write("\n4. Meat")
menu_file.write("\n5. Onions")
menu_file.write("\n6. Corn")
menu_file.write("\n")
menu_file.close()

#menu.txt dosyasını okur
menu_file = open("/content/menu.txt", "r")
menu = menu_file.read()
menu_file.close()

print(menu) #menüyü ekrana yazdırır
print("Please answer your pizza and sauce choices with the numbers on the menu!")

hello = input("Choose a pizza base: ") #kullanıcı pizza tabanı tercihini yapar

#pizza tabanı fiyatları
pizza_base = dict()
pizza_base['Pizza capricciosa'] = 85
pizza_base['Pizza Margherita'] = 80
pizza_base['Hawaiian pizza'] = 82
pizza_base['Seafood pizza'] = 87

#sos fiyatları
sauce = dict()
sauce['Olives'] = 3
sauce['Mushrooms'] = 3.5
sauce['Goat Cheese'] = 5
sauce['Meat'] = 7
sauce['Onions'] = 2.5
sauce['Corn'] = 4

class pizza(object): #pizza üst sınıfı
  def get_description(self): #açıklamalar
    if hello == "1":
      return "Similar to Pizza quattro stagioni, but with toppings mixed rather\nthan separated.\nIngredients: Ham, mushrooms, artichokes, egg"
    elif hello == "2":
      return "The archetypical Neapolitan pizza.\nIngredients: Tomatoes, mozzarella, basil"
    elif hello == "3":
      return "Tends to divide opinion.\nIngredients: Pineapple, ham"
    elif hello == "4":
      return "Subvarieties include Pizza frutti di mare (no cheese) and Pizza\npescatore (with mussels or squid).\nIngredients: Seafood, such as fish, shellfish or squid"    
    else:
      return "nah"

  def get_cost(self): #fiyatlar
    if hello == "1":
      return pizza_base["Pizza capricciosa"]
    elif hello == "2":
      return pizza_base["Pizza Margherita"]
    elif hello == "3":
      return pizza_base["Hawaiian pizza"]
    elif hello == "4":
      return pizza_base["Seafood pizza"]    
    else:
      return "nah"

choice = pizza()

#pizza sınıfının alt sınıfları
class Pizza_capricciosa(pizza):
  def Pizza_capricciosa1(self): 
    return choice.get_description()
  def Pizza_capricciosa2(self): 
    return choice.get_cost()

class Pizza_Margherita(pizza):
  def Pizza_Margherita1(self): 
    return choice.get_description()
  def Pizza_Margherita2(self): 
    return choice.get_cost()

class Hawaiian_pizza(pizza):
  def Hawaiian_pizza1(self): 
    return choice.get_description()
  def Hawaiian_pizza2(self): 
    return choice.get_cost()

class Seafood_pizza(pizza):
  def Seafood_pizza1(self): 
    return choice.get_description()
  def Seafood_pizza2(self): 
    return choice.get_cost()

x1 = Pizza_capricciosa()
x2 = Pizza_Margherita()
x3 = Hawaiian_pizza()
x4 = Seafood_pizza()

def selected_one(): #açıklama ve fiyat bilgisini kullanıcı onaylamadan önce ekrana bastıran fonksiyon
  if hello == "1":
    x = x1.Pizza_capricciosa1()
    y = x1.Pizza_capricciosa2()
    return f"{x}\nPrice: {y} tl"
  elif hello == "2":
    x = x2.Pizza_Margherita1()
    y = x2.Pizza_Margherita2()
    return f"{x}\nPrice: {y} tl"
  elif hello == "3":
    x = x3.Hawaiian_pizza1()
    y = x3.Hawaiian_pizza2()
    return f"{x}\nPrice: {y} tl"
  elif hello == "4":
    x = x4.Seafood_pizza1()
    y = x4.Seafood_pizza2()
    return f"{x}\nPrice: {y} tl" 
  else:
    return "nah"  

while True: #kullanıcının pizza seçimini onaylamadığında tekrar tercih yapmasını sağlar
  print(selected_one())
  hello2 = input("Do you confirm your choice?(Please only answer 'yes' or 'no'): ")
  if hello2 == "yes":
    then = input("Choose a pizza sauce: ")
    break
  elif hello2 == "no":
    hello = input("Choose a pizza base: ")
    
def base_price(): #pizza fiyat fonksiyonu
  if hello == "1":
    y = x1.Pizza_capricciosa2()
    return y
  elif hello == "2":
    y = x2.Pizza_Margherita2()
    return y
  elif hello == "3":
    y = x3.Hawaiian_pizza2()
    return y
  elif hello == "4":
    y = x4.Seafood_pizza2()
    return y 
  else:
    return "nah" 

class decorator(object): #sos sınıfı
  def get_cost(self): #fiyatlar
    if then == "1":
      return sauce["Olives"]
    elif then == "2":
      return sauce["Mushrooms"]
    elif then == "3":
      return sauce["Goat Cheese"]
    elif then == "4":
      return sauce["Meat"]
    elif then == "5":
      return sauce["Onions"]    
    elif then == "6":
      return sauce["Corn"]        
    else:
      return "nah"

choice2 = decorator()

#sosların alt sınıfları
class olives(decorator):
  def olive_sauce(self):
    return choice2.get_cost()

class mushrooms(decorator):
  def mushroom_sauce(self):
    return choice2.get_cost()

class goat_cheese(decorator):
  def goat_cheese_sauce(self):
    return choice2.get_cost()

class meat(decorator):
  def meat_sauce(self):
    return choice2.get_cost()

class onions(decorator):
  def onion_sauce(self):
    return choice2.get_cost()

class corn(decorator):
  def corn_sauce(self):
    return choice2.get_cost()

y1 = olives()
y2 = mushrooms()
y3 = goat_cheese()
y4 = meat()
y5 = onions()
y6 = corn()

def selected_two(): #sos fiyat bilgisini ekrana yazdırır
  if then == "1":
    return f"Price: {y1.olive_sauce()} tl"
  elif then == "2":
    return f"Price: {y2.mushroom_sauce()} tl"
  elif then == "3":
    return f"Price: {y3.goat_cheese_sauce()} tl"  
  elif then == "4":
    return f"Price: {y4.meat_sauce()} tl"
  elif then == "5":
    return f"Price: {y5.onion_sauce()} tl" 
  elif then == "6":
    return f"Price: {y6.corn_sauce()} tl"  
  else:
    return "nah" 

while True: #kullanıcının pizza sosu onaylamadığında tekrar tercih yapmasını sağlar
  print(selected_two())
  then2 = input("Do you confirm your choice?(Please only answer 'yes' or 'no'): ")
  if then2 == "yes":
    print("\nPayment(Only credit/debit card payment)")
    break
  elif then2 == "no":
    then = input("Choose a pizza sauce: ")

def sauce_price(): #sos fiyat fonksiyonu
  if then == "1":
    y = y1.olive_sauce()
    return y
  elif then == "2":
    y = y2.mushroom_sauce()
    return y
  elif then == "3":
    y = y3.goat_cheese_sauce()
    return y
  elif then == "4":
    y = y4.meat_sauce()
    return y 
  elif then == "5":
    y = y5.onion_sauce()
    return y
  elif then == "6":
    y = y6.corn_sauce()
    return y   
  else:
    return "nah" 

#pizza ve sos fiyatını toplar ve ekrana fiyat bilgisini yazdırır
total_price = base_price() + sauce_price()
print("Total price:", total_price, "tl")

#kullacı ödeme kısmına geçer
user_name = input("User name: ")
user_id = input("User id: ")
credit_card_number = input("Credit card number: ")
credit_card_password = input("Credit card password: ")

if hello == "1":
  a = "Pizza capricciosa"
elif hello == "2":
  a = "Pizza Margherita"
elif hello == "3":
  a = "Hawaiian pizza"
elif hello == "4":
  a = "Seafood pizza"  
else:
  print("nah")

if then == "1":
  b = "Olives"
elif then == "2":
  b = "Mushrooms"
elif then == "3":
  b = "Goat cheese"
elif then == "4":
  b = "Meat"
elif then == "5":
  b = "Onions"
elif then == "6":
  b = "Corn"  
else:
  print("nah")     

#alınan bilgiler Orders_Database.csv tablosunda tutulur
title = ["User name", "User id", "Credit card number", "Credit card password", "Pizza order", "Description of order", "Time order"]
data = [user_name, user_id, credit_card_number, credit_card_password, f"{a} with {b}", choice.get_description(), 
        datetime.datetime.now()]

with open("Orders_Database.csv", "w") as table:
  x = csv.writer(table)
  x.writerow(title)
  x.writerow(data)

```
